```python
import pandas as pd
import mplfinance as mpf

# Sample data: Dates, Open, High, Low, Close prices
data = {
    'Date': [
        '2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04',
        '2024-10-05', '2024-10-06', '2024-10-07', '2024-10-08'
    ],
    'Open': [100, 102, 105, 103, 107, 109, 111, 114],
    'High': [105, 106, 107, 108, 110, 112, 115, 118],
    'Low': [95, 101, 100, 102, 104, 106, 110, 113],
    'Close': [102, 105, 103, 107, 109, 111, 114, 116]
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])  # Convert the 'Date' column to datetime
df.set_index('Date', inplace=True)  # Set the 'Date' column as the index

# Create the candlestick chart
mpf.plot(df, type='candle', style='charles', title='Candlestick Chart',
         ylabel='Price', volume=False)
```
